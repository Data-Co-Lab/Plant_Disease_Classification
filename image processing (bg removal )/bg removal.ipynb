```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
def preprocess_grabcut(path): 
    original_image = cv2.imread(path)
    original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
    original_image = cv2.resize(original_image , (256,256))
    image = original_image
    # create a simple mask image similar to the loaded image, with theshape and return type 
    mask = np.zeros(image.shape[:2], np.uint8) 

    # specify the background and foreground model 
    backgroundModel = np.zeros((1, 65), np.float64) 
    foregroundModel = np.zeros((1, 65), np.float64) 

    # define the Region of Interest (ROI) as the coordinates of the rectangle 
    rectangle = (20, 20, 200, 200) 

    # apply the grabcut algorithm with appropriate 
    cv2.grabCut(image, mask, rectangle,   
                backgroundModel, foregroundModel, 
                1, 
                cv2.GC_INIT_WITH_RECT) 

    # In the new mask image, pixels will be marked with four flags  
    # four flags denote the background / foreground  
    # mask is changed, all the 0 and 2 pixels  
    # are converted to the background 
    # mask is changed, all the 1 and 3 pixels 
    # are now the part of the foreground 
    # the return type is also mentioned, 
    # this gives us the final mask 
    mask2 = np.where((mask == 2)|(mask == 0), 0, 1).astype('uint8') 

    # The final mask is multiplied with the input image to give the segmented image. 
    image = image * mask2[:, :, np.newaxis] 
    
    f = plt.figure()
    f.add_subplot(1,2, 1)
    plt.imshow(image)
    f.add_subplot(1,2, 2)
    plt.imshow(original_image)
    plt.show(block=True)
    #plt.imshow(original_image),plt.imshow(image)
```

```python
for i in range(1,12):
    preprocess_grabcut('test_images/'+str(i)+'.jpg')
```
